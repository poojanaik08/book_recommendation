# Popularity Based Recommendation System

Highest Average Rating Books (Top 50 books): Criteria: Considering Books with minimum 250 votes

```python
#merge ratings on top of books on the ISBN column
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'}, inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
popular_df
```

```python
#consider books whose num_ratings is greater than 250
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings', ascending=False).head(20)
popular_df
```

```python
popular_df = popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Year-Of-Publication', 'Image-URL-M', 'num_ratings', 'avg_ratings']]
```

# Colaborative Based Recommendation System

```python
#merged df of ratings and books
ratings_with_name
```

```python
#considering users who have rated minimum 200 books 
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
#boolean indexing
x[x]
```

```python
freq_users = x[x].index
freq_users
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(freq_users)]
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values= 'Book-Rating')
pt
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

Finding similar books using eucledian distance.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score
```

```python
similarity_score.shape
```

```python
#creating a function that takes the name of a book and gives 5 suggestions that are similar to the book
def recommend(book_name):
    #index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1], reverse=True)[1:6]
    data = []
    for i in similar_items:
        item = []
        tem_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(tem_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(tem_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(tem_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    return data
```

```python
recommend('The Da Vinci Code')
```
